# StressDream / Vista — Demo

Runs noise steering on the bundled `example_images/truck.jpg` and compares:
- **With regularizer** (default): iter 1 (nominal baseline) vs. iter 20 (steered)
- **Without regularizer**: iter 1 vs. iter 20 — ablation to see the typical-set constraint's effect

> **Timing:** a single Vista generation takes ~1–2 min on an H100.  
> With 20 iterations the full optimization run takes ~30–40 min.

Run from the `StressDream/` directory.

```python
import os, sys
if os.path.basename(os.getcwd()) == 'vista':
    os.chdir('..')
sys.path.insert(0, os.getcwd())

from IPython.display import Video, display
import glob

def show(path, width=900):
    display(Video(path, embed=True, width=width))

def latest_run(base_dir):
    """Return the most recently created timestamped subdirectory."""
    runs = sorted(glob.glob(os.path.join(base_dir, '*')))
    return runs[-1] if runs else None

print('Working dir:', os.getcwd())
```

## Run 1 — With regularizer (default config)

```python
# ~30-40 min on H100 (20 iters × ~1-2 min each)
!python vista/run_steering.py
```

```python
run_dir = latest_run('vista/outputs/vista_steering')
print('Run dir:', run_dir)

print('\n--- Iter 1 (before optimization) ---')
show(os.path.join(run_dir, 'iter_0001.mp4'))

print('\n--- Iter 20 (steered) ---')
show(os.path.join(run_dir, 'iter_0020.mp4'))
```

## Run 2 — Without regularizer (ablation)

```python
# ~30-40 min on H100
!python vista/run_steering.py --no_regularizer
```

```python
run_dir_noreg = latest_run('vista/outputs/vista_steering')
print('Run dir (no-reg):', run_dir_noreg)

print('\n--- No-reg · Iter 1 (before optimization) ---')
show(os.path.join(run_dir_noreg, 'iter_0001.mp4'))

print('\n--- No-reg · Iter 20 (steered) ---')
show(os.path.join(run_dir_noreg, 'iter_0020.mp4'))
```

## Side-by-side comparison

| | Iter 1 (baseline) | Iter 20 (steered) |
|---|---|---|
| **With regularizer** | `run_dir/iter_0001.mp4` | `run_dir/iter_0020.mp4` |
| **No regularizer** | `run_dir_noreg/iter_0001.mp4` | `run_dir_noreg/iter_0020.mp4` |

The regularizer keeps the noise in the Gaussian typical set, preventing the optimizer from collapsing to adversarial/degenerate samples.
